```python
import matplotlib.pyplot as plt
```

```python
def pull_data(file):
    ans = -1
    with open(file, 'r') as f:
        ans = float(f.readline().strip())
    return ans
```

```python
probability_ticks = [0, 1, 2, 4, 6, 8]
probability = [0, 1e-9, 2e-9, 4e-9, 6e-9, 8e-9]
tests = ["high_school_computer_science", "astronomy"]
data = [[pull_data(f"results/gpt2_{str(prob)}_{test}") for prob in probability] for test in tests]
```

```python
fig, ax = plt.subplots(1, 2)

fig.suptitle("GPT2 Tests")

fig.text(0.5, 0.02, 'Error Probability (1e-9)', ha='center')
fig.text(0.04, 0.5, 'Score', va='center', rotation='vertical')

plt.setp(ax, xlim=(-1, 9), ylim=(0, 0.5), xticks=[0, 2, 4, 6, 8])

ax[0].set_title("Computer Science")
ax[0].scatter(probability_ticks, data[0])

ax[1].set_title("Astronomy")
ax[1].scatter(probability_ticks, data[1])
```
